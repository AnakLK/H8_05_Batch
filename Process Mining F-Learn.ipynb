```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('event log.csv')
```

```python
data.info()
```

```python
data.drop('IP address', axis=1, inplace=True)
data.drop('Origin', axis=1, inplace=True)
data.drop('Event context', axis=1, inplace=True)
data.drop('Component', axis=1, inplace=True)
data.drop('Affected user', axis=1, inplace=True)
```

```python
data
```

```python
data = data[data['User full name'] != 'Hanita Yulia']
data = data[data['User full name'] != 'Flexible Learning UKSW']
data = data[data['User full name'] != '-']

# Menghilangkan aktivitas pengajar, sistem, dan tidak ada nama
```

```python
unique_values = data['Event name'].unique()

print("\nJenis nilai di kolom 'Course':")
print(unique_values)

# Mengecek isi event apa saja di dalam kolom event name
```

```python
data = data[data['Event name'] != 'User enrolled in course']
data = data[data['Event name'] != 'Role assigned']
data = data[data['Event name'] != 'Submission updated.']
data = data[data['Event name'] != 'Grade overview report viewed']
data = data[data['Event name'] != 'Badge listing viewed']
data = data[data['Event name'] != 'Comment created']
data = data[data['Event name'] != 'Remove submission confirmation viewed.']
data = data[data['Event name'] != 'User profile viewed']
data = data[data['Event name'] != 'Lesson restarted']
data = data[data['Event name'] != 'The status of the submission has been updated.']
data = data[data['Event name'] != 'User unenrolled']
data = data[data['Event name'] != 'Tour started']
data = data[data['Event name'] != 'Tour ended']
data = data[data['Event name'] != 'Role unassigned']
data = data[data['Event name'] != 'Step shown']
data = data[data['Event name'] != 'Comment deleted']
data = data[data['Event name'] != 'Course user report viewed']

# Menghapus event-event yang tidak terlalu diperlukan
```
